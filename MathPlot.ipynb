```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import math
```

```python
list1=["2021-08-24","2021-08-25","2021-08-26","2021-08-27","2021-08-28"]
list2=[180, 186, 176, 200, 210]
xpoints=np.array(list1)
ypoints=np.array(list2)
plt.plot(xpoints,ypoints, color='g',marker='*',linewidth=1)
plt.title("Date vs Deaths")
plt.xlabel("Date")
plt.ylabel("No of Deaths")
plt.grid(True, color='b')
plt.show()
```

```python
plt.scatter(xpoints,ypoints)
plt.show()
```

```python
plt.bar(xpoints,ypoints)
plt.show()
```

```python
plt.pie(ypoints,labels=xpoints)
plt.show()
```
