```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
list1=["Day1","Day2","Day3","Day4","Day5"]
list2=[100,165,128,190,198]
```

```python
xpoints=np.array(list1)
ypoints=np.array(list2)
```

```python
#plt.plot(xpoints,ypoints)
#plt.plot(xpoints,ypoints, 'r')
plt.title("Day vs Cases")
plt.xlabel("Day")
plt.ylabel("Cases")
plt.grid()
plt.plot(xpoints,ypoints, color='b',marker='o',linestyle="solid")
plt.show()
```

```python
plt.scatter(xpoints,ypoints,color='g')
plt.show()
```

```python
plt.bar(xpoints,ypoints,color='c')
plt.show()
```

```python
plt.pie(ypoints,labels=xpoints)
plt.show()
```
