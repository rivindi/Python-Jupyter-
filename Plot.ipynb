```python
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
list1=[1,2,3,4,5]
list2=[181,176,189,190,200]
xpoints=np.array(list1)
ypoints=np.array(list2)
```

```python
plt.plot(xpoints,ypoints,linestyle='solid',marker='o',color='g')
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Function")
plt.grid()
plt.show()
```

```python
plt.scatter(xpoints,ypoints,color='g')
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Function")
plt.grid()
plt.show()
```
