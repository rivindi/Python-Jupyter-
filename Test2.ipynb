```python
import matplotlib.pyplot as plt
import numpy as np
import math
list1=np.arange(-10,11,1)
list2=[]
def cal(x):
    y=x**3-3*x-2
    return y

for x in list1:
    y=cal(x)
    list2.append(y)
print(list2)
```

```python
plt.xlabel("X")
plt.ylabel("Y")
plt.plot(list1,list2,'r')
plt.show()
```

```python
list3=np.arange(0,2*math.pi,math.pi/6)
list4=[]
def cal(x):
    y=math.sin(x)**2+x+1
    return y

for x in list3:
    y=cal(x)
    list4.append(y)

arr1=np.array(list4)
print(arr1)
```

```python
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Function")
plt.plot(list3,list4,'ro')
plt.show()
```
