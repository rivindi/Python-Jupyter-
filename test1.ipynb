```python
import math
#f(x0)
def getF(x):
    fx=math.sin(x)*math.sin(x)-x**2+1
    return fx
def getFD(x):
    fdx=2*math.sin(x)*math.cos(x)-2*x
    return fdx
def rCal(x):
    #x1=x0-f(x0)/f'(x0)
    x1=x-getF(x)/getFD(x)
    return round(x1,6)
i=0
x=math.pi
while i<6:
    x=rCal(x)
    i=i+1
    print(x)
```

```python
import math
#f(x0)
def getF(x):
    fx=math.log(math.sin(x)*math.sin(x))+x**2
    return fx
def getFD(x):
    fdx=2/math.tan(x)+2*x
    return fdx
def rCal(x):
    #x1=x0-f(x0)/f'(x0)
    x1=x-getF(x)/getFD(x)
    return round(x1,6)
i=0
x=1
while i<6:
    x=rCal(x)
    i=i+1
    print(x)
```
